Step 4: Exploring Data

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/mcdonalds.csv')  # Replace 'your_dataset.csv' with the actual file path

# Display variable names
variable_names = data.columns.tolist()
print("Variable Names:", variable_names)

# Display sample size
sample_size = len(data)
print("Sample Size:", sample_size)

# Display the first three rows of the data
first_three_rows = data.head(3)
print("First Three Rows of Data:\n", first_three_rows)
```

```python
import numpy as np

# Select columns 1 to 11
MD_x = data.iloc[:, 0:11]
MD_like = data.iloc[-4]

# Convert "Yes" to 1 and "No" to 0
MD_x = (MD_x == "Yes").astype(int)

# Calculate column means
col_means = np.round(MD_x.mean(), 2)

print("Column Means:\n", col_means)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
MD_x_standardized = scaler.fit_transform(MD_x)

# Perform PCA
pca = PCA()
MD_pca = pca.fit(MD_x_standardized)

# Summary of the PCA
print("Explained Variance Ratio:\n", MD_pca.explained_variance_ratio_)
print("\nPrincipal Components:\n", MD_pca.components_)
```

```python
print("PCA Results:")
print("Standard deviations (eigenvalues):\n", MD_pca.explained_variance_)
print("\nProportional variance:\n", MD_pca.explained_variance_ratio_)
print("\nCumulative variance:\n", np.cumsum(MD_pca.explained_variance_ratio_))
print("\nRotation (principal components):\n", MD_pca.components_)
print("\nMean:\n", MD_pca.mean_)
print("\nTransformed data:\n", MD_pca.transform(MD_x_standardized))
```

```python
import matplotlib.pyplot as plt

# Project the data onto the principal components
projected_data = MD_pca.transform(MD_x_standardized)

# Plotting the projected data
plt.scatter(projected_data[:, 0], projected_data[:, 1], c="grey")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Projected Data")
plt.show()

# Get the projection axes
projection_axes = MD_pca.components_

print("Projection Axes (Principal Components):\n", projection_axes)
```

Step 5: Extracting Segments

5(a)
-
Using k-Means

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Determine the number of clusters using the silhouette score
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(projected_data)
    silhouette_scores.append(silhouette_score(projected_data, cluster_labels))

# Plot silhouette scores
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs. Number of Clusters")
plt.show()
```

```python
num_clusters = 4  # Choose an appropriate number of clusters

# Perform KMeans clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(projected_data)

# Add cluster labels to the dataframe
data['Cluster'] = cluster_labels

# Display the number of samples in each cluster
cluster_counts = data['Cluster'].value_counts()
print("Number of Samples in Each Cluster:\n", cluster_counts)
```

```python
# Plot clusters on the projected data plot
for cluster in range(num_clusters):
    cluster_data = projected_data[cluster_labels == cluster]
    plt.scatter(cluster_data[:, 0], cluster_data[:, 1], label=f"Cluster {cluster}")

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Projected Data with Clusters")
plt.legend()
plt.show()
```

```python
# Plot histograms for each cluster
for cluster in range(num_clusters):
    cluster_data = projected_data[cluster_labels == cluster]

    plt.figure(figsize=(8, 6))
    plt.hist(cluster_data[:, 0], bins=20, alpha=0.5, label=f"Cluster {cluster} (PC1)")
    plt.hist(cluster_data[:, 1], bins=20, alpha=0.5, label=f"Cluster {cluster} (PC2)")

    plt.xlabel("Principal Component Values")
    plt.ylabel("Frequency")
    plt.title(f"Distribution of Principal Component Values for Cluster {cluster}")
    plt.legend()
    plt.show()
```

```python
import numpy as np
from sklearn.metrics import pairwise_distances
import matplotlib.pyplot as plt

# Calculate the pairwise distances between segments
pairwise_dist = pairwise_distances(MD_x.T, metric='hamming')

# Calculate segment level stability within solutions
MD_r4 = 1 - pairwise_dist

# Plot segment level stability within solutions
plt.figure(figsize=(8, 6))
plt.plot(MD_r4, marker='o')
plt.xlabel("Segment Number")
plt.ylabel("Segment Stability")
plt.title("Segment Level Stability within Solutions (SLSW) Plot")
plt.ylim(0, 1)
plt.show()
```

5(b)
-
Using Mixtures of Distributions

```python
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans

# Initialize Gaussian Mixture Model
n_components_range = range(2, 9)
bic_scores = []
aic_scores = []
icl_scores = []

for n_components in n_components_range:
    gmm = GaussianMixture(n_components=n_components, covariance_type='full', n_init=10, random_state=1234)
    gmm.fit(MD_x)
    bic_scores.append(gmm.bic(MD_x))
    aic_scores.append(gmm.aic(MD_x))
    # Calculate ICL (Integrated Completed Likelihood)
    icl = gmm.score_samples(MD_x).sum() - gmm.lower_bound_
    icl_scores.append(icl)

# Plot information criteria
plt.figure(figsize=(8, 6))
plt.plot(n_components_range, aic_scores, label='AIC')
plt.plot(n_components_range, bic_scores, label='BIC')
plt.plot(n_components_range, icl_scores, label='ICL')
plt.xlabel('Number of Components')
plt.ylabel('Information Criteria')
plt.title('Information Criteria for Gaussian Mixture Model')
plt.legend()
plt.show()

# Fit Gaussian Mixture Model with the selected number of components
n_components = 4
gmm_final = GaussianMixture(n_components=n_components, covariance_type='full', n_init=10, random_state=1234)
clusters_gmm = gmm_final.fit_predict(MD_x)

# Perform K-means clustering
kmeans = KMeans(n_clusters=n_components, random_state=1234)
clusters_kmeans = kmeans.fit_predict(MD_x)

# Compare GMM clusters with K-means clusters
table = np.zeros((n_components, n_components), dtype=int)
for i in range(n_components):
    for j in range(n_components):
        table[i, j] = np.sum((clusters_kmeans == i) & (clusters_gmm == j))

print("Cross-tabulation of K-means clusters and GMM clusters:\n", table)
```
